# The sigmoid and its derivative

```python
x = np.linspace(-10,10,50)
```

```python
np.exp(-1*x)/(1. + np.exp(x))**2
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
sig = 1./(1.+np.exp(-x))
sigp = np.exp(-x)/(1.+np.exp(-x))**2

plt.plot(x,sig)
plt.plot(x,sigp,color = 'red')
plt.title('The sigmoid function and its derivative')
plt.xlabel('Time (arbitrary units)')
plt.ylabel('Fraction of total deaths')
plt.legend(['sigmoid: cumulative deaths','derivative: daily deaths'],loc='upper left')
plt.grid()
```
